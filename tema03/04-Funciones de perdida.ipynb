```python
import tensorflow as tf
import matplotlib.pyplot as plt #para la representacion grafica
session = tf.Session()
```

```python
#Generamos 500 valores entre -1. y 1.
x_vals = tf.linspace(-1.,1.,500)
target = tf.constant(0.)
```

## Tipos de funciones de pérdida en problemas de predicción (REGRESIÓN)

## Norma en L2 (distancia euclidea)
$$ L2(y_r, y_p) =  (y_r - y_p)^2 $$
- Tiende a alejarse del valor en las outlayers

```python
l2_y_vals = tf.square(target-x_vals)
```

```python
l2_y_out = session.run(l2_y_vals)
```

```python
#Utilizando la funcion de nn
l2_y_out2= tf.nn.l2_loss(target-x_vals)
```

```python
print(l2_y_out2)
```

## Norma L1 (Valor Absoluto)
$$ L1(y_r, y_p) =  |y_r - y_p| $$

- Mejor que L2 para outlayer xq no están al cuadrado los valores.
- No es derivable. Puede ir a algoritmos queno converjan.

```python
l1_y_vals = tf.abs(target-x_vals)
```

```python
l1_y_out =session.run(l1_y_vals)
```

```python
print(l1_y_out)
```

## Pseudo de Huber
Toma lo menor de la norma L1 y L2
No sacrifica los outlayers
Continua y derivable
Convexa cerca del valor objetivo, menos valores extremos
Depende de un parámetro extra delta.

$$ PH(y_r,y_p) = \delta^2 \cdot (\sqrt{1+(\frac{y_r-y_p}{\delta}})^2 -1)  ,  \delta > 0 $$

```python
delta1 = tf.constant(0.25)
phuber1_y_vals = tf.multiply(tf.square(delta1),tf.sqrt(1.+tf.square((target-x_vals)/delta1))-1.)
```

```python
phuber1_y_out = session.run(phuber1_y_vals)
print(phuber1_y_out)
```

```python
delta2 = tf.constant(5.)
phuber2_y_vals = tf.multiply(tf.square(delta2),tf.sqrt(1.+tf.square((target-x_vals)/delta2))-1.)
phuber2_y_out = session.run(phuber2_y_vals)
print(phuber2_y_out)
```

# Tipos de funciones de pérdidas para problemas de clasificación.

```python
x_vals = tf.linspace(-3., 5., 500)
target = tf.constant(1.)
targets = tf.fill([500,], 1.)
```

# Hinge (función bisagra)

### Calcular las perdidas entre las 2 posibles clases de clasificación 1 y -1
### En el ejemplo Evaluar que tan cerca del 1 está nuestra predicción

$$  H(y_r,y_p) = max(0, 1-y_r\cdot y_p) $$

```python
hinge_y_vals = tf.maximum(0.,1. - tf.multiply(target, x_vals))
hinge_y_out = session.run(hinge_y_vals)
print(hinge_y_out)
```

## Entropía cruzada (función logística)
Muy utilizada en clasificación binaria.
Medir la distancia de la clase actual con la predicción.

$$ H(y_r,y_p) = -y_r \cdot log(y_p)- (1-y_r)\cdot log(1-y_p) $$

```python
xentr_y_vals = -tf.multiply(target, tf.log(x_vals))-tf.multiply(1. -target, tf.log(1. -x_vals))
xentr_y_out = session.run(xentr_y_vals)
print(xentr_y_out)
```

# Entroía cruzada de Sigmoide

Se aplica la sigmoidea antes de aplicar la fórmula de la extropía cruzada.

$$ H(y_r,y_p) = -y_r \cdot log(sigmoide(y_p))- (1-y_r)\cdot log(1-sigmoide(y_p)) $$

logits= predicciones= x_vals


```python
xentr_sig_y_vals = tf.nn.sigmoid_cross_entropy_with_logits(logits=x_vals,labels=targets)
xentr_sig_y_out = session.run(xentr_sig_y_vals)
print(xentr_sig_y_out)
```

```python
#Versión ponderada

pos_weight = tf.constant(0.5)
xentr_sig_w_y_vals = tf.nn.weighted_cross_entropy_with_logits(logits=x_vals, targets=targets, pos_weight=pos_weight)
xentr_sig_w_y_out = session.run(xentr_sig_w_y_vals)
print(xentr_sig_w_y_out)
```

## Softmax Cross Entropy 
Para valores que no están normalizados.
Utilizada cuando hay una sola categoria a predecir.
$$ SM(z_k) = \frac{e^{z_k}}{\sum_{i=1}^n e^{z_i}} $$

```python
unscaled_logits = tf.constant([[1.,-3.,10.]]) #categorias z_k
target_dist = tf.constant([[0.1,0.02,0.88]])  #probabilidades de pertenecer a las categorias
softmax_xentr = tf.nn.softmax_cross_entropy_with_logits_v2(logits=unscaled_logits, labels=target_dist)
softmax_xentr_y_out = session.run(softmax_xentr)
print(softmax_xentr_y_out)
```

## Sparse softmax cross entropy
target tiene un vector sparse donde tiene un 1 (un indice)
target no es probabilidad

```python
unscaled_logits = tf.constant([[1.,-3.,10.]]) #categorias z_k
sparsed_target_dist = tf.constant([2])  #probabilidades de pertenecer a las categorias
sparsed_xentr = tf.nn.sparse_softmax_cross_entropy_with_logits(logits=unscaled_logits, labels=sparsed_target_dist)
sparsed_xentr_y_out = session.run(sparsed_xentr)
print(sparsed_xentr_y_out)
```
